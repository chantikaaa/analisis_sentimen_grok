```python
# import library
from google_play_scraper import reviews_all, Sort
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scrape all reviews for the 'Grok' app in playstore (penilaian dalam bahasa Inggris)
app_id = 'ai.x.grok'
max_comments = 11000

reviews = reviews_all(
    app_id,
    lang='en',  # English reviews only
    sort=Sort.MOST_RELEVANT
)

# Limit review count
limited_reviews = reviews[:max_comments]
```

```python
df = pd.DataFrame(limited_reviews)
print(f"Total jumlah review yang diambil: {len(df)}")
```

```python
# Save to CSV format
df.to_csv('grok_reviews_limited.csv', index=False)
```
